```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import Data
lifesat = pd.read_csv("data/lifesat/lifesat.csv")


X= lifesat[["GDP per capita (USD)"]].values
Y = lifesat[["Life satisfaction"]].values
fig, ax = plt.subplots()
fig = plt.scatter(X, Y)
ax.set_xlabel("GDP per capita (USD)")
ax.set_ylabel("Life satisfaction")
ax.grid()
plt.axis([23500, 62500, 4, 9])

from sklearn import linear_model
model = linear_model.LinearRegression()

model.fit(X,Y)

intercept = model.intercept_
coefficient = model.coef_

ax.plot(X, X * coefficient + intercept, color = "red")


from sklearn import neighbors
model_neighbor = neighbors.KNeighborsRegressor(n_neighbors= 3)

# Train the model 
model_neighbor.fit(X, Y)



plt.show()
```

<h3> Model Prediction for Cyprus

```python
X_new = [[37655.2]]   #Cyprus GDP per capita in 2020
model.predict(X_new)
```

```python
from sklearn import neighbors
model_neighbor = neighbors.KNeighborsRegressor(n_neighbors= 3)

# Train the model 
model_neighbor.fit(X, Y)
model_neighbor.predict(X_new)
```
